```python
df.head()
```

```python
# Cleaning code from the last lab:

df.rename(columns= {"st" : "state"}, inplace=True)
df
```

```python
# Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
print(df.gender.unique())
df.gender = df.gender.replace({'F': 'F', 'M': 'M', 'Femal': 'F', 'Male': 'M', 'female': 'F', 'male': 'M'})
```

```python
display(df)
```

```python
# - *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
print(df.state.unique())
df.state = df.state.replace({"AZ": "Arizona", "Cali": "California", "WA": "Washington"})


```

```python
# - *In education, "Bachelors" could be replaced by "Bachelor"*
print(df.education.unique())
df.education = df.education.replace({"Bachelors": "Bachelor"})
print(df.education.unique())
```

```python
df.customer_lifetime_value = df.customer_lifetime_value.str.rstrip('%').astype('float')
print(df.customer_lifetime_value)
```

```python
# - *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*
print(df.vehicle_class.unique())
df.vehicle_class = df.vehicle_class.replace({"Sports Car": "Luxury", "Luxury SUV": "Luxury", "Luxury Car": "Luxury"})
```

```python
# *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. 
# As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. 
# Finally, since it should be numeric, cast the column to be in its proper type.*

df.number_of_open_complaints = df.number_of_open_complaints.str.split("/").str[1].astype(float)
# df.number_of_open_complaints = df.number_of_open_complaints.astype(float)
df.number_of_open_complaints.unique()
```

```python
# Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.
null_counts = df.isnull().sum()
display(null_counts)
```

```python
df = df.dropna(how="all")  # Removes rows with null values
```

```python
df.describe()
```

```python
df.gender.unique()
df.gender = df.gender.fillna('U')
```

```python
null_counts = df.isnull().sum()
display(null_counts)
```

```python
df.customer_lifetime_value.unique()
```

```python
df.customer_lifetime_value = df.customer_lifetime_value.fillna(df.customer_lifetime_value.mean())
```

```python
df.loc[:, 'customer_lifetime_value'] = df['customer_lifetime_value'].fillna(df['customer_lifetime_value'].mean())
```

```python
# Your code goes here
df4.head()
```

```python
pivot_df4 = df4.pivot_table(index='sales_channel', values='total_claim_amount', aggfunc='sum')
display(pivot_df4)
```

### Insights

The pivot table shows that the sales channel 'Agent' brought in the most sales in terms of total revenue, followed by 'Branch', 'Call Center', and 'Web'. This indicates that the 'Agent' channel is the most effective in generating revenue, while the 'Web' channel is the least effective.
